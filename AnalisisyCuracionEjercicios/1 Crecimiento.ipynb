```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
fig, (ax_peso, ax_peso2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 6), sharex=True)
fig.suptitle('Distribucion de Peso en Percentiles' )
# peso
percentiles_peso = pd.read_csv('wtageinf.csv', index_col='Agemos')
percentiles_peso.query('Sex==1')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(ax=ax_peso, cmap='jet',  title ='Peso Nios', linewidth=5, alpha=0.5)
percentiles_peso2 = percentiles_peso
percentiles_peso2.query('Sex==2')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(ax=ax_peso2, cmap='jet',  title = "Peso Nias", linewidth=5, alpha=0.5)
plt.show()
```

```python
fig, (ax_altura, ax_altura2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 6), sharex=True)
fig.suptitle('Distribucion de Peso en Percentiles' )
# peso
percentiles_altura= pd.read_csv('lenageinf2.csv' ,  index_col='Agemos')
percentiles_altura.query('Sex==1')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(ax=ax_altura, cmap='jet',  title ='Altura Nios', linewidth=5, alpha=0.5)
percentiles_altura2 = percentiles_altura
percentiles_altura2.query('Sex==2')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(ax=ax_altura2, cmap='jet',  title = "Altura Nias", linewidth=5, alpha=0.5)
plt.show()

```
